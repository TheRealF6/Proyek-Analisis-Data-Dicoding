### Cleaning Data

**Mengubah format "dteday" di day.csv dari object menjadi datetime**

```python
datetime_columns = ["dteday"]

for column in datetime_columns:
  day_df[column] = pd.to_datetime(day_df[column])
```

**Mengubah format "dteday" di hour.csv dari object menjadi datetime**

```python
datetime_columns = ["dteday"]

for column in datetime_columns:
  hour_df[column] = pd.to_datetime(hour_df[column])
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Tren penggunaan sepeda berbagi
day_df['dteday'] = pd.to_datetime(day_df['dteday'])

plt.figure(figsize=(15,5))
sns.lineplot(x='dteday', y='cnt', data=day_df)

plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

plt.title('Tren Penggunaan Bike Sharing (2011 - 2012)')
plt.xlabel('Bulan dan Tahun')
plt.ylabel('Jumlah Penyewaan')

plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Pertanyaan 2:

```python
# Perbandingan antara hari kerja dan akhir pekan
plt.figure(figsize=(12,6))
sns.boxplot(x='workingday', y='cnt', data=day_df, width=0.5)

medians = day_df.groupby(['workingday'])['cnt'].median().values
median_labels = [str(np.round(s, 2)) for s in medians]
pos = range(len(medians))
for tick, label in zip(pos, median_labels):
    plt.text(pos[tick], medians[tick] + 100, median_labels[tick],
             horizontalalignment='center', size='x-small', color='black', weight='semibold')

plt.title('Perbandingan Penggunaan Bike Sharing Antara Hari Kerja dan Akhir Pekan')
plt.xlabel('Akhir Pekan (0) vs Hari Kerja (1)')
plt.ylabel('Jumlah Penyewaan')
plt.grid(True)
plt.show()
```

## Conclusion

### **Pertanyaan 1**
- Terlihat bahwa waktu di mana terjadi fluktuasi penggunaaan bike sharing terjadi ketika memasuki bulan Maret/April yang kebetulan sudah memasuki musim semi yang berarti banyak orang yang menggunakan transportasi ini untuk menuju tempat kerja atau sekolah.
- Terjadi penurunan saat akhir tahun dikarenakan masa libur dan awal dari musim dingin.
- Tren tertinggi terjadi sekitar bulan September tahun 2012 (musim gugur).

### **Pertanyaan 2**
- Pada hari kerja jumlah minimal dan maksimal pengguna bike sharing jauh lebih banyak dibandingkan akhir pekan.
- Juga bisa dilihat bahwa pada akhir pekan jumlah pengguna lebih beragam dibandingkan hari kerja.
- Median juga lebih tinggi pada hari kerja.
- Kesimpulannya secara umum lebih banyak yang menggunakan bike sharing pada hari kerja dibandingkan akhir pekan.
